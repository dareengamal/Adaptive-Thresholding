Import Libraries and Load the Image:

1-Import the necessary libraries: cv2, numpy, and matplotlib.

2-Load the input image using OpenCV's cv2.imread() function.

3- Convert the image to grayscale for thresholding.

```python
# Import necessary libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load input image
img = cv2.imread('practice.jpg')

#convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img,cmap='gray')
```

Apply Global Thresholding:

1-Apply global thresholding (cv2.threshold()) to the grayscale image.

2-Comment on the results, discussing its limitations in handling varying lighting conditions.

```python
# Apply global thresholding
ret, global_thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

plt.imshow(global_thresh, cmap='gray')
plt.title('Global Thresholding')
plt.show()
```

```python
# Comment on the results
# Global thresholding works well when the lighting conditions are relatively consistent across the image.
# However, when there are areas of the image with varying lighting conditions, it can lead to over or under segmentation.
```

#Apply Adaptive Thresholding:

1-Experiment with different adaptive thresholding methods using cv.
adaptiveThreshold(), and comment on the results difference between adaptive and global thresholding:

- cv2.ADAPTIVE_THRESH_MEAN_C

- cv2.ADAPTIVE_THRESH_GAUSSIAN_C

2-Varying Block Sizes:

- Apply adaptive thresholding with different constant values (e.g., 5, 10, 20). And comment on the results for each case

```python
# Experiment with adaptive thresholding methods
# 1- Mean thresholding
mean_thresh = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)

# Display mean thresholding result
plt.imshow(mean_thresh, cmap='gray')
plt.title('Mean Thresholding')
plt.show()
```

```python
# Gaussian thresholding
gaussian_thresh = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)

# Display Gaussian thresholding result
plt.imshow(gaussian_thresh, cmap='gray')
plt.title('Gaussian Thresholding')
plt.show()
```

```python
# Comment on the results above
# adaptive thresholding is more effective than global thresholding in handling varying lighting conditions.


# Mean thresholding works well for images with relatively uniform lighting conditions, while Gaussian thresholding is more effective
#for images with varying lighting .
```

```python
# Varying block sizes
# Apply adaptive thresholding with different block sizes
constant = 5
block_size = 11

thresh1 = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, block_size, constant)
thresh2 = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, block_size, constant)

constant = 10

thresh3 = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, block_size, constant)
thresh4 = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, block_size, constant)

constant = 20

thresh5 = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, block_size, constant)
thresh6 = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, block_size, constant)
```

```python

# Display thresholding results
plt.figure(figsize=(15, 10))

plt.subplot(231)
plt.imshow(thresh1, cmap='gray')
plt.title('Block Size: 11, Constant: 5')

plt.subplot(232)
plt.imshow(thresh2, cmap='gray')
plt.title('Block Size: 11, Constant: 5')

plt.subplot(233)
plt.imshow(thresh3, cmap='gray')
plt.title('Block Size: 11, Constant: 10')

plt.subplot(234)
plt.imshow(thresh4, cmap='gray')
plt.title('Block Size: 11, Constant: 10')

plt.subplot(235)
plt.imshow(thresh5, cmap='gray')
plt.title('Block Size: 11, Constant: 20')

plt.subplot(236)
plt.imshow(thresh6, cmap='gray')
plt.title('Block Size: 11, Constant: 20')

plt.show()
```

```python
# Comment
# The choice of block size and constant value can have a significant impact on the thresholding result.
# Smaller block sizes and higher constant values results in more detailed thresholding and vice versa.
```
